# Valeurs manquantes / doublons

```python
df.isnull().sum()
```

```python
doublons = df.duplicated()
df
```

```python
# Utiliser value_counts pour chaque colonne et afficher les doublons
for column in df.columns:
    print(f"Doublons dans la colonne '{column}':")
    compte_values = df[column].value_counts()
    
    # Afficher les valeurs qui apparaissent plus d'une fois (doublons)
    doublons = compte_values[compte_values > 1]
    
    if not doublons.empty:
        print(doublons)
    else:
        print("Aucun doublon trouvé.")
    print('-' * 50)
```
